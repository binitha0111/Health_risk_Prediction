```sql
-- Risk summary by readmission status
SELECT 
    readmitted,
    COUNT(*) as patient_count,
    ROUND(AVG(time_in_hospital), 2) as avg_days_in_hospital,
    ROUND(AVG(num_lab_procedures), 2) as avg_lab_procedures,
    ROUND(AVG(num_medications), 2) as avg_medications,
    ROUND(AVG(num_active_meds), 2) as avg_active_meds,
    ROUND(AVG(total_prior_visits), 2) as avg_prior_visits
FROM workspace.default.reporting_diabetic
GROUP BY readmitted
ORDER BY patient_count DESC
```
