```sql
-- High risk indicators summary
SELECT
    CASE WHEN on_insulin = 1 THEN 'On Insulin' ELSE 'No Insulin' END as insulin_status,
    CASE WHEN high_utilizer = 1 THEN 'High Utilizer' ELSE 'Normal' END as utilizer_status,
    COUNT(*) as patient_count,
    ROUND(AVG(readmitted_binary) * 100, 1) as readmission_rate_pct,
    ROUND(AVG(num_lab_procedures), 1) as avg_lab_procedures,
    ROUND(AVG(time_in_hospital), 1) as avg_days_in_hospital
FROM workspace.default.reporting_diabetic
GROUP BY on_insulin, high_utilizer
ORDER BY readmission_rate_pct DESC
```
