```sql
-- Patient count by age group and readmission
SELECT 
    age,
    COUNT(*) as total_patients,
    SUM(readmitted_binary) as readmitted_count,
    ROUND(SUM(readmitted_binary) * 100.0 / COUNT(*), 1) as readmission_rate_pct
FROM workspace.default.reporting_diabetic
GROUP BY age
ORDER BY age
```
